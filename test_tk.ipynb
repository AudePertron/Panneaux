```python
model = tf.keras.models.load_model('model/model_classification')

class GUI(Frame):

    def __init__(self, master=None):
        Frame.__init__(self, master)
        w,h = 650, 650
        master.minsize(width=w, height=h)
        master.maxsize(width=w, height=h)
        self.pack()

        self.file = Button(self, text='Parcourir', command=self.choose)
        self.choose = Label(self, text="Choisir une photo").pack()
        
        #Replace with your image
        self.image = PhotoImage(file='logo.png')
        self.label = Label(image=self.image)
        self.pred_label = Label()
        
        self.file.pack()
        self.label.pack()
        self.pred_label.pack()
        
    def choose(self):
        ifile =  filedialog.askopenfilename(initialdir = "/",title = "Select file",filetypes = (("jpeg files",".jpg"),("all files",".*")))
        #ifile = filedialog.askopenfile(parent=self,mode='rb',title='Choisir un fichier')
        path = Image.open(ifile)
        print(ifile)
        self.image2 = ImageTk.PhotoImage(path)
        self.label.configure(image=self.image2)
        self.label.image=self.image2
        self.prediction_image(ifile)


    def prediction_image(self, path):
        global pred_label 
        img_blurr = cv2.imread(path)

        img_reshaped = cv2.resize(img_blurr,(23, 23), interpolation = cv2.INTER_LINEAR)

        x_int = np.array(img_reshaped)
        x_int.shape
        x_int= x_int/ 255

        x_int = np.reshape(x_int,(-1,23,23,3))
        prediction = (model.predict(x_int))
        
        idx = np.argmax(prediction)
        pourcentage = round(prediction[0][idx]*100,3)
        
        #affichage de la prédiction
        self.pred_label.configure(text = "Prédiction : " +str(categories[idx]) +"\n Certitude : " + str(pourcentage))
        
root = Tk()
app = GUI(master=root)
app.mainloop()
root.destroy()
```
